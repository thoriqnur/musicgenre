```python
import pandas as pd
import numpy as np
import time
import pickle as pkl

from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.naive_bayes import MultinomialNB


from learn.classifier import NaiveBayes
from learn.metrics import accuracy, confusion_matrix
```

```python
df = pd.read_csv('./data/genres.csv') #import data dari final genres (hanya 4 genres)
print(df.shape) 
df.head()
```

```python
X = df.drop(columns="Target") #semua data kecuali kolom target
y = df["Target"] #hanya colom target
print(X.shape, y.shape)
# print(X, y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
nb_sklearn = MultinomialNB()

min_max_scaler = preprocessing.MinMaxScaler()
X_train_minmax = min_max_scaler.fit_transform(X_train)
t_sklearn_start = time.time()
nb_sklearn.fit(X_train_minmax, y_train)
t_sklearn = time.time() - t_sklearn_start

y_pred_sklearn = nb_sklearn.predict(X_test)

acc_sklearn = round(accuracy(y_test, y_pred_sklearn) * 100, 3)

# print("Accuracy using sklearn package\t\t: {0}%, ({1} seconds)".format(acc_sklearn, t_sklearn))
# matrix, label = confusion_matrix(y_test, y_pred_sklearn, show_label=True)
# print(matrix)
# print(label)
```
